```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
def cost(y, ttx):
    z = 1.0 / (1 + np.exp (-ttx))
    return -y * np.log (z) - (1- y) * np.log (1 - z)
```

```python
ttx = np.linspace (-3, 3, 100)

fig, ax = plt.subplots (nrows=1, ncols=1, figsize = (12, 6))
colors = ['r', 'b']
for y in [0, 1]:
    ax.plot (ttx, cost(y, ttx), color=colors[y], label='$y = {}$'.format (y))

ax.set_xlabel (r'$\theta^{T}x$', fontsize=20)
ax.set_ylabel (r'$J(\theta)$', fontsize=20)
plt.legend ()
plt.grid (True)
```
